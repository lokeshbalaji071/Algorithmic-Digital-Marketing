```python
def get_splited_df_dict(df: 'pd.DataFrame', split_column: 'str'):
    """
    splits a pandas.DataFrame on split_column and returns it as a dict
    """

    df_dict = {value: df[df[split_column] == value] for value in df[split_column].unique()}

    return df_dict
```

```python
splitted = get_splited_df_dict(df, 'product')
# # accessing the DataFrame with 'some_column_value'
# splitted[some_column_value] 
```

```python
prod_list = df['product'].unique().tolist()
```

```python
prod_list
```

```python
len(prod_list)
```

```python
review_list = pd.DataFrame(prod_list, columns =['Products'])
```

```python
review_list.head()
```

```python
def get_count(prodname):
    prod_df = splitted[prodname] 
    pos_count = ((prod_df['Analysis']).value_counts()).tolist()
    return pos_count
```

```python
pos_count = []
for prod in prod_list:
    pos_cmt = get_count(prod)
    pos_count.append(pos_cmt)
```

```python
pos_count
```

```python
comm_df = pd.DataFrame(pos_count, columns = ['Positive', 'Neutral', 'Negative'])
comm_df['Negative'] = comm_df['Negative'].fillna(0)
comm_df = comm_df.astype({"Negative":'int'})
comm_df.head()
```

```python
comm_df['Negative'] += comm_df['Neutral']
comm_df.head()
```

```python
comm_df.drop('Neutral', axis=1, inplace=True)
comm_df.head()
```

```python
review_list = pd.concat([review_list, comm_df], axis=1)
```
